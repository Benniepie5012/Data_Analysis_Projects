```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_white"
import kaleido 

data = pd.read_csv("supply_chain_data.csv")
data.head()
```

```python
# Define the number of decimal places to round to
#decimal_places = 2

# Round up the values in the 'Values' column to the specified decimal places
#df['Values'] = df['Values'].round(decimal_places)
```

```python
decimal_places = 4
data['Manufacturing costs'] = data['Manufacturing costs'].round(decimal_places)
```

```python
decimal_places = 4
data['Costs'] = data['Costs'].round(decimal_places)

decimal_places = 4
data['Revenue generated'] = data['Revenue generated'].round(decimal_places)

decimal_places = 4
data['Price'] = data['Price'].round(decimal_places)

decimal_places = 4
data['Defect rates'] = data['Defect rates'].round(decimal_places)
```

```python
data.head()
```

```python
print(data.describe())
```

```python
fig = px.scatter(data, x='Price', y='Revenue generated',
                color='Product type', hover_data=['Number of products sold'], 
                 trendline="ols") 
fig.show("png")
```

```python
data.head()
```

```python
sales_data = data.groupby('Product type')['Number of products sold'].sum().reset_index()

pie_chart = px.pie(sales_data, values= 'Number of products sold',
                  names='Product type', title= 'Sales by Product Type',
                  hover_data = ['Number of products sold'], hole =0.5,
                  color_discrete_sequence=px.colors.qualitative.Pastel)

pie_chart.show("png")
```

```python
data.tail()
```

```python
total_revenue = data.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()

bar_chart = px.bar(total_revenue, x='Shipping carriers', y='Revenue generated',
                   title='Total Revenue by Shipping Carrier')
                

bar_chart.show("png")
```

```python
avg_lead_time = data.groupby('Product type')['Lead time'].mean().reset_index()

avg_manufacturing_costs = data.groupby('Product type')['Manufacturing costs'].mean().reset_index()

result = pd.merge(avg_lead_time, avg_manufacturing_costs, on='Product type')

result.rename(columns={'Lead time': 'Average Lead Time', 'Manufacturing costs': 'Average Manufacturing Costs'}, inplace=True)
print(result)
```

```python
revenue_chart = px.line(data, x='SKU', 
                        y='Revenue generated', 
                        title='Revenue Generated by SKU')
revenue_chart.show("png")
```

```python
stock_chart = px.line(data, x='SKU', 
                      y='Stock levels', 
                      title='Stock Levels by SKU')
stock_chart.show("png")
```

```python
order_quantity_chart = px.bar(data, x='SKU', 
                              y='Order quantities', 
                              title='Order Quantity by SKU')
order_quantity_chart.show("png")
```

```python
shipping_cost_chart = px.bar(data, x='Shipping carriers', 
                             y='Shipping costs', 
                             title='Shipping Costs by Carrier')
shipping_cost_chart.show('png')
```

```python
transportation_charts = px.pie(data, 
                               values="Costs", names="Transportation modes",
                               title='Cost Distribution by Transportation mode', 
                               hole=0.5,
                               color_discrete_sequence =px.colors.qualitative.Pastel)

transportation_charts.show("png")
```

```python
defect_rates_by_product = data.groupby('Product type')['Defect rates'].mean().reset_index()

fig = px.bar(defect_rates_by_product, x='Product type', y='Defect rates',
             title='Average Defect Rates by Product Type')
fig.show("png")
```

```python
pivot_table = pd.pivot_table(data, values='Defect rates', 
                             index=['Transportation modes'], 
                             aggfunc='mean')
print(pivot_table)
```

```python
transportation_chart = px.pie(values=pivot_table["Defect rates"], 
                              names=pivot_table.index, 
                              title='Defect Rates by Transportation Mode',
                              hole=0.5,
                              color_discrete_sequence=px.colors.qualitative.Pastel)

transportation_chart.show('png')
```
